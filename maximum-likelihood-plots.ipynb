```python
import matplotlib.pyplot as plt
import numpy as np
import operator
```

```python
xx = np.arange(0., 1.01, 0.01)
```

```python
def plot_coins(ax, flips):
    results = flips.split()
    C_H = operator.countOf(results, "H")
    C_T = len(results) - C_H
    ax.plot(xx, [x**C_H * (1-x)**C_T for x in xx], label = flips)
```

```python
fig, ax = plt.subplots()

plot_coins(ax, "H H T H")
plot_coins(ax, "H T T T")
plot_coins(ax, "H T T H")
plot_coins(ax, "H H H H")

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.set_ylim([0,0.2])


# plt.show()
plt.savefig('all.pdf', format='pdf')
```
